In this notebook, we generate strain history data.

### General Libraries

```python
import numpy as np
from data_generator import StrainHistoryGenerator
```

```python
# Configuration
n_histories = 1000
n_steps     = 101
eps_max     = 0.03
theta_max   = np.pi / 6
seed        = 42            # For reproducibility

output_dir  = "./strain_histories"
```

```python
generator = StrainHistoryGenerator(
    n_steps=n_steps,
    eps_max=eps_max,
    theta_max=theta_max,
    seed=seed
)
```

```python
histories = generator.generate_multiple_histories(n_histories=n_histories)
```

```python
generator.save_histories( histories=histories, 
                    prefix = "strain",
                    directory = output_dir,
                    include_cartesian=True)
```

```python
generator.plot_distribution(
                    histories=histories,
                    bins=50,
                    )
```

```python
generator.plot_strain_ratio_distribution(histories=histories,
                                    bins = 50)
```

```python
generator.plot_j2_distribution(histories=histories,
                                    bins = 50)
```

```python
generator.plot_example()
```
